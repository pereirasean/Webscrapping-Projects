```python
from bs4 import BeautifulSoup
import requests
 
url = "http://example.com/"
 
# Getting the webpage, creating a Response object.
response = requests.get(url)
 
# Extracting the source code of the page.
data = response.content
print(data)
 
# Passing the source code to Beautiful Soup to create a BeautifulSoup object for it.
soup = BeautifulSoup(data, 'html.parser')
 
# Extracting all the <a> tags into a list.
tags = soup.find_all('a')
 
# Extracting URLs from the attribute href in the <a> tags.
for tag in tags:
    print(tag.get('href'))
```

```python
import re

#regrex=re.compile('[abc]')
#regrex=re.compile('a')
#regrex=re.compile('[a-zA-Z]')
#regrex=re.compile('[^a-zA-Z]')
regrex=re.compile('[+]')



#regrex.match('A')
regrex.match('+')
```

```python
#Matches any decimal digit (0-9)

regrex=re.compile('\d')

#Matches any non-decimal digit (^0-9)

regrex=re.compile('\D')


#Matches any white space character

regrex=re.compile('\s')

#Matches any non-white space character

regrex=re.compile('\S')

#Matches any aplhanumer space character [a-zA-Z0-9_]

regrex=re.compile('\w')

#Matches any non-aplhanumer  space character [^a-zA-Z0-9_]

regrex=re.compile('\W')
```

```python
#* is infinite times (0,inf)
#regrex=re.compile('a*')
regrex=re.compile('[a-c]*')

#regrex.match('aaaa')
regrex.match('aaaaaaaaaaaaaaaac')
```

```python
#+: character can be matched once or more time: such it should come atlest once

#regrex=re.compile('a+')
regrex=re.compile('[a-c]+')
regrex.match('aaaaaaaaaaaaaaaa')
```

```python
#? says the previous character can come once or none

regrex=re.compile('a?b')

regrex.match('ab')
```

```python
#{m,n} : the qualifier mentions that there should be atleast m repititon and at most n


regrex=re.compile('a{2,4}')
regrex.match('aaaaabaaaaaaa')
```

```python
#as many values i.e. inf for second position, if first-value is blank means =0
regrex=re.compile('a{2,}')

#^ says the string should start woth the first letter
regrex=re.compile('^abc')

#| says the string should have either both of the character
regrex=re.compile('a|b')

#$ says the string should end woth the last letter
regrex=re.compile('abc$')
```
